```python
from docx import Document
from docx.shared import Pt
import os
import re

def sanitize_filename(text):
    """Dosya adlarında geçersiz karakterleri temizle"""
    return re.sub(r'[\\/*?:"<>|]', "", text.replace(" ", "_"))[:60]

def split_docx_by_multiple_font_sizes(docx_path, output_folder, target_font_sizes_pt=(25, 19, 16)):
    os.makedirs(output_folder, exist_ok=True)
    doc = Document(docx_path)

    sections = {}
    current_heading = "BASLANGIC"
    idx = 1

    for para in doc.paragraphs:
        text = para.text.strip()
        if not text:
            continue

        found_target = False
        for run in para.runs:
            if run.font.size and run.font.size.pt in target_font_sizes_pt:
                found_target = True
                break

        if found_target:
            # Yeni başlık tespit edildi
            clean_title = sanitize_filename(text)
            current_heading = f"{idx:02d}_{clean_title}"
            sections[current_heading] = ""
            idx += 1
        else:
            sections.setdefault(current_heading, "")
            sections[current_heading] += para.text + "\n"

    # Kaydet
    for title, content in sections.items():
        filepath = os.path.join(output_folder, f"{title}.txt")
        with open(filepath, "w", encoding="utf-8") as f:
            f.write(content)

    print(f"✅ Belge başarıyla {len(sections)} bölüme ayrıldı. Klasör: {output_folder}")

# === Kullanım Örneği ===
split_docx_by_multiple_font_sizes(
    "ornek_donusturulmus.docx", 
    "docx_bolunmus_25_19_16pt", 
    target_font_sizes_pt=(25, 19, 16)
)
```

```python
from together import Together

def ask_together(content):

    client = Together(api_key=?)

    response = client.chat.completions.create(
    model="meta-llama/Llama-4-Maverick-17B-128E-Instruct-FP8",
    messages=[{"role":"user","content":content}],
    temperature=0.51,
    top_p=0.91,
    seed=None
)
    return response.choices[0].message.content
```

```python
import os

def promp_soru(content, filename):
    prompt = """
Aşağıda sana verdiğim ingilizce paragrafı birebir Türkçeye çevir. Tamamen aynı çeviri olması lazım.
Eğer Python kodları varsa, satır satır açıklama ver.
Başlıkları markdown'a uygun şekilde # başlık olarak yap.
Ayırca kodları düzegün olarak tekrar yaz ve aynısı olsun
"""

    full_prompt = prompt + content + "\n"

    # Kayıt yolu (güncel klasör ismi kullanabilirsin)
    md_file = os.path.join(filename + ".md")

    # Önceden varsa temizle
    if os.path.exists(md_file):
        os.remove(md_file)

    # API çağrısı
    result = ask_together(full_prompt)
    if result is None:
        print(f"Hata: API'den boş yanıt alındı.")
        result = "HATA: API yanıtı alınamadı."

    # Kaydet
    with open(md_file, "a", encoding="utf-8") as f:
        f.write(result)
        f.write("\n\n---\n\n")

    print(f"✅ '{md_file}' dosyası oluşturuldu.")
```
